### Part-02: Lab Task is to visualize the datasets or data using Matplotlib builtin function Graphs

```python
people = ['goku', 'luffy','natsu', 'zeref']
age =[25, 30, 35, 40, 45]
height =[145, 151, 165, 173]
weight=[45, 55, 65, 75]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(weight, height)
plt.title("Relationship between Weight and Height of Patients")
plt.ylabel("Height in Meters")
plt.xlabel("Weight of Patients in Kgs")
plt.show()
```

### Generate Bar Plot

```python
plt.bar(people, height)
plt.xlabel('People')
plt.ylabel('Height')
plt.title('Height of Patiants')
plt.show()
```

### Generate Histogram

```python
plt.hist(age, bins=6, color='purple', edgecolor ='black')
plt.xlabel('Age')
plt.ylabel('No. of Patients')
plt.title('Age of Patients')
plt.show()
```
